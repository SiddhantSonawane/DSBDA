```python
import requests
from bs4 import BeautifulSoup as bs
import numpy as np
import pandas as pd
```

```python
link = "https://www.amazon.in/2022-Apple-MacBook-Laptop-chip/product-reviews/B0B3C6FBPF/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
page = requests.get(link)
page
```

```python
soup = bs(page.content, 'html.parser')
print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-profile-name')
names
```

```python
cust_names = []

for i in range(len(names)):
    cust_names.append(names[i].get_text())
    
cust_names
```

```python
len(cust_names)
```

```python
titles = soup.find_all('a', class_='review-title-content')
titles
```

```python
review_titles =[]

for i in range(len(titles)):
    review_titles.append(titles[i].get_text())
    
review_titles
```

```python
review_titles[:] = [titles.rstrip('\n') for titles in review_titles]
review_titles
```

```python
ratings = soup.find_all('i', class_='review-rating')
ratings
```

```python
review_ratings = []

for i in range(0, len(ratings)):
    review_ratings.append(ratings[i].get_text())
    
review_ratings
```

```python
reviews = soup.find_all('span', {"data-hook":"review-body"})
reviews
```

```python
review_content =[]

for i in range(0, len(reviews)):
    review_content.append(reviews[i].get_text())

review_content
```

```python
review_content[:] = [reviews.strip("\n") for reviews in review_content]
review_content[:] = [reviews.rstrip("\n") for reviews in review_content]
review_content
```

```python
df = pd.DataFrame()
df['Customer Name'] = cust_names[:10]
df['Review Title'] = review_titles[:10]
df['Review Rating'] = review_ratings[:10]
df['Review Content'] = review_content[:10]
df
```

```python
df.to_csv(r'./reviews.csv',index=True)
```
